```python
import pandas as pd
import numpy as np

# Create dataset
data = {
    "Name": ["Rahul", "Amit", "Neha", "Riya", "Karan",
             "Rahul", "Simran", "Priya", "Arjun", "Vikas"],

    "Branch": ["CSE", "ECE", "CSE", "ECE", "MECH",
               "CSE", "MECH", "ECE", "CSE", "MECH"],

    "Marks": [85, 78, np.nan, 70, 88,
              85, 90, np.nan, 95, 60],

    "Age": [22, 25, 21, 23, np.nan,
            22, 24, 25, 22, np.nan]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("students_lab.csv", index=False)

print(df)
```

```python
import pandas as pd
raw_data = [
    ['Alice', 'Engineering', 'Senior', 110000, 15000],
    ['Bob', 'Marketing', 'Junior', 70000, 5000],
    ['Charlie', 'Engineering', 'Senior', 100000, 12000],
    ['David', 'Marketing', 'Senior', 95000, 8000],
    ['Eva', 'Engineering', 'Junior', 85000, 6000],
    ['Frank', 'Marketing', 'Junior', 65000, 4000]
]
df2 = pd.DataFrame(raw_data)
```

```python
df2
```

```python
columns = ["Employee", "Team", "Level", "Salary", "Bonus"]
df2 = pd.DataFrame(raw_data, columns=columns)
df2
```

```python
df2['Team_Avg_Salary'] = df2.groupby('Team')['Salary'].transform('mean')
df2
```
